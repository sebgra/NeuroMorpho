# L-Measure function for feature extraction

```python
def get_measurment(neuron):
    
    features = dict()
    features['Surface'] = getOneMeasure('Surface',neuron)
    features['Soma_Surface'] = getOneMeasure('Soma_Surface',neuron)
    features['N_stems']= getOneMeasure('N_stems',neuron)
    features['N_bifs']= getOneMeasure('N_bifs',neuron)
    features['N_branch']= getOneMeasure('N_branch',neuron)
    features['N_tips']= getOneMeasure('N_tips',neuron)
    features['Width']= getOneMeasure('Width',neuron)
    features['Height']= getOneMeasure('Height',neuron)
    features['Depth']= getOneMeasure('Depth',neuron)
    features['Type']= getOneMeasure('Type',neuron)
    features['Diameter']= getOneMeasure('Diameter',neuron)
    features['Diameter_pow']= getOneMeasure('Diameter_pow',neuron)
    features['Length']= getOneMeasure('Length',neuron)
    features['Surface']= getOneMeasure('Surface',neuron)
    features['SectionArea']= getOneMeasure('SectionArea',neuron)
    features['Volume']= getOneMeasure('Volume',neuron)
    features['EucDistance']= getOneMeasure('EucDistance',neuron)
    features['PathDistance']= getOneMeasure('PathDistance',neuron)
    features['Branch_Order']= getOneMeasure('Branch_Order',neuron)
    features['Terminal_degree']= getOneMeasure('Terminal_degree',neuron)
    features['TerminalSegment']= getOneMeasure('TerminalSegment',neuron)
    features['Taper_1']= getOneMeasure('Taper_1',neuron)
    features['Taper_2']= getOneMeasure('Taper_2',neuron)
    features['Branch_pathlength']= getOneMeasure('Branch_pathlength',neuron)
    features['Contraction']= getOneMeasure('Contraction',neuron)
    features['Fragmentation']= getOneMeasure('Fragmentation',neuron)
    features['Daughter_Ratio']= getOneMeasure('Daughter_Ratio',neuron)
    features['Parent_Daughter_Ratio']= getOneMeasure('Parent_Daughter_Ratio',neuron)
    features['Partition_asymmetry']= getOneMeasure('Partition_asymmetry',neuron)
    features['Rall_Power']= getOneMeasure('Rall_Power',neuron)
    features['Pk']= getOneMeasure('Pk',neuron)
    features['Pk_classic']= getOneMeasure('Pk_classic',neuron)
    features['Pk_2']= getOneMeasure('Pk_2',neuron)
    features['Bif_ampl_local']= getOneMeasure('Bif_ampl_local',neuron)
    features['Bif_ampl_remote']= getOneMeasure('Bif_ampl_remote',neuron)
    features['Bif_tilt_local']= getOneMeasure('Bif_tilt_local',neuron)
    features['Bif_tilt_remote']= getOneMeasure('Bif_tilt_remote',neuron)
    features['Bif_torque_local']= getOneMeasure('Bif_torque_local',neuron)
    features['Bif_torque_remote']= getOneMeasure('Bif_torque_remote',neuron)
    features['Last_parent_diam']= getOneMeasure('Last_parent_diam',neuron)
    features['Diam_threshold']= getOneMeasure('Diam_threshold',neuron)
    features['HillmanThreshold']= getOneMeasure('HillmanThreshold',neuron)
    features['Helix']= getOneMeasure('Helix',neuron)
    features['Fractal_Dim']= getOneMeasure('Fractal_Dim',neuron)
    
    return features
```

# Statistics on distances

```python
def distance_stats(distribution) : 
    
    statistics = dict()
    
    statistics['mean'] = np.mean(distribution)
    statistics['median'] = np.median(distribution)
    statistics['std'] = np.std(distribution)
    statistics['var'] = np.var(distribution)
    statistics['min'] = np.min(distribution)
    statistics['max'] = np.max(distribution)
    statistics['skew'] = stats.describe(distribution)[-2]
    statistics['kurtosis'] = stats.describe(distribution)[-1]
    statistics['D1'] = np.percentile(distribution,10)
    statistics['D2'] = np.percentile(distribution,20)
    statistics['D3'] = np.percentile(distribution,30)
    statistics['D4'] = np.percentile(distribution,40)
    statistics['D5'] = np.percentile(distribution,50)
    statistics['D6'] = np.percentile(distribution,60)
    statistics['D7'] = np.percentile(distribution,70)
    statistics['D8'] = np.percentile(distribution,80)
    statistics['D9'] = np.percentile(distribution,90)
    
    return statistics
```

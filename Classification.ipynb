```python
import numpy as np
from pylmeasure import *
import glob
import os.path
import pandas as pd

import matplotlib.pyplot as plt
```

```python
def LM_dict_to_pd(LM_dict) : 
    
    features_type  = list(LM_dict.item().keys())

    colnames = list()
    
    val = list()

    for f in features_type:
    
        sub_feature = list(LM_dict.item()[f].keys())
        
        val = val + list(LM_dict.item()[f].values())

    
        #print(sub_feature)
    
        for sf in sub_feature:
    
    #colnames.append(str(str(f) + '_' + str(sub_feature)))
    
            colnames.append(str(f+'_'+sf))
        
    df = pd.DataFrame([val],columns=colnames)
        
    return df


def Stat_dict_to_pd(Stat_dict):
    
    
    df = pd.DataFrame([Stat_dict.item().values()],columns=Stat_dict.item().keys())
    
    return df


def merge_df(df_LM,df_Stats) : 
    
    df_binded = pd.concat([df_LM.reset_index(drop=True), df_Stats], axis=1)
    return df_binded
```

```python
Basket_tag = np.repeat(0,50)
Motoneuron_tag = np.repeat(1,75)
Purkinje_tag = np.repeat(2,120)
Pyramidal_tag = np.repeat(3,141)
Retina_tag = np.repeat(4,189)

labels = np.concatenate((Basket_tag,Motoneuron_tag))
labels = np.concatenate((labels,Purkinje_tag))
labels = np.concatenate((labels,Pyramidal_tag))
labels = np.concatenate((labels,Retina_tag))

labels
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_np, labels, test_size=0.2, random_state=0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=20, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
